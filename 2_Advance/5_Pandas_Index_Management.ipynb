```python
#importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#data cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.index
```

```python
df_india = df[df['job_country']=='India']

df_india.head()
```

```python
df_india.reset_index(inplace=True)
df_india.head()
```

```python
df_india.set_index('index', inplace=True)
```

```python
df_india
```

```python
median_pivot = df_india.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])

median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=True)
```

```python
median_pivot.sort_index(inplace=True)
median_pivot
```
