```python
date_count = df.job_posted_date.value_counts()
date_count = date_count.sort_index()
plt.plot(date_count.index, date_count.values)
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
monthly_count = df.job_posted_month.value_counts()
monthly_count = monthly_count.sort_index()
plt.plot(monthly_count.index, monthly_count.values)
plt.show()
```

```python
jobs_count = df.job_title_short.value_counts()
jobs_count = jobs_count.sort_values(ascending=True)
plt.title("Job Details", fontsize=16)
plt.xlabel("Job Title", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.bar(jobs_count.index, jobs_count.values)
plt.xticks(rotation=40, ha='right')
plt.show()
```

```python
df.plot(x='job_posted_date', y='salary_year_avg', kind='bar')
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
job_salary.plot(kind='barh')
plt.xlabel("Salary in $")
plt.ylabel("")
plt.title("Median Job Salary")
plt.show()
```
